# Bar Plot

```python
from matplotlib import pyplot as plt
```

```python
import numpy as np
import pandas as pd
```

```python
x=['Python','Java','C','Swift']
y=[89,85,48,79]
z=[23,54,37,50]  
width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Languages',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.title('Data Visualization',fontsize=16)

plt.bar(p,y,width,color='orange',alpha=0.7,edgecolor='black',label='Data')
plt.bar(p1,z,width,color='pink',alpha=0.7,edgecolor='black',label='Data 1')
plt.xticks(p+width/2,x,rotation=22)
plt.legend()
plt.show()
```

```python
x=['Python','Java','C','Swift']
y=[89,85,48,79]
z=[23,54,37,50]  
width=0.4
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel('Languages',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.title('Data Visualization',fontsize=16)

plt.barh(p,y,width,color='orange',alpha=0.7,edgecolor='black',label='Data')
plt.barh(p1,z,width,color='pink',alpha=0.7,edgecolor='black',label='Data 1')
plt.xticks(p+width/2,x,rotation=22)

plt.legend()
plt.show()
```

# Scatter Plot

```python
xx=[1,2,3,4,5,6,7]
yy=[1,3,2,4,2,5,4]
no=[1,3,2,4,6,3,4]
#crs=['red','g','b','y','g','r','m']
crs=[10,33,22,44,55,22,33]
sizes=[40,35,73,111,45,31,65]
plt.scatter(xx,yy,c=crs,s=sizes,marker='.',cmap='Accent')
plt.scatter(xx,no,c=crs,s=sizes,marker='.',cmap='Blues_r')

plt.title('Scatter Plot',fontsize=15)
plt.xlabel('Day',fontsize=15)
t=plt.colorbar()
t.set_label('ColorBar')
plt.ylabel('No',fontsize=15)
plt.show()
```

# Bar Plot

```python
x1=['Basanta','Anuragh','Abdulla','subash','Deepak']
y1=[78,78,80,56,60]
z=[85,74,66,68,47]
width=0.4
p=np.arange(len(x1))
p1=[j+width for j in p] 

plt.bar(p,y1,width,color='Orange',alpha=0.8,edgecolor='green',linewidth=0.6,linestyle='dashed',label='Marks 1st sem')
plt.bar(p1,z,width,color='grey',alpha=0.8,edgecolor='green',linewidth=0.6,linestyle='dashed',label='Marks 2nd sem')
plt.xticks(p+width/2,x1)
plt.legend()
plt.xlabel('Students',fontsize=15)
plt.ylabel('Marks',fontsize=15)
plt.title('Computer Engineering')
plt.show()
```

# Pie Plot

```python
pie=[5,15,25,30,12,13]
y=['Go','C','Java','Python','PHP','Javascript']
ex=[0.0,0.2,0.0,0.0,0.0,0.0]
plt.pie(pie,labels=y,explode=ex,autopct='%0.1f%%',shadow=True,radius=1,startangle=20,textprops={'fontsize':12},wedgeprops={'linewidth':2,'edgecolor':'darkgray'})
plt.title('Data Show')
plt.legend(loc=2)
plt.show()
```

# Stem Plot

```python
p=[1,2,3,4,5,6]
q=[2,5,7,3,8,1]
plt.stem(p,q,linefmt=':',markerfmt='ro',bottom=6,basefmt='g',label='Python',use_line_collection=False,orientation='horizontal')
plt.legend()
plt.show()
```

# Box Plot & Whisker Plot

```python
x=[10,20,30,40,50,60]
y=[60,50,40,30,20,10]
plt.boxplot([x,y],notch=False,vert=True,widths=0.2,labels=['Python','C++'],patch_artist=False,showmeans=True,whis=3.2)

plt.show()
```

# Stack Plot

```python
ar=[1,2,3,4,5]
ary=[3,2,1,4,30]
ary1=[4,12,10,4,1]
ary2=[3,22,11,4,10]
l=['Area','Area1','Area2','Area3',]
plt.stackplot(ar,ary,ary1,ary2,labels=l,baseline='wiggle')
plt.legend(loc=2)
plt.show()
```

# Step Plot

```python
x=[1,2,3,4,5]
y=[1,2,3,4,5]
plt.step(x,y,marker='.',ms=10,mfc='blue',label='Python',color='red')
plt.grid()
plt.title('Python')
plt.xlabel('X- axis')
plt.ylabel('Y- axis')
plt.legend()
plt.show()
```

# Fill-Between Plot

```python
xq=[1,2,3,4,5]
yq=[1,2,3,4,5]
plt.plot(xq,yq,color='r')
plt.xlabel('X- axis')
plt.ylabel('Y- axis')
plt.fill_between(xq,yq)
plt.title('Python')
plt.show()
```
