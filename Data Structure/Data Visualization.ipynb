```python
from matplotlib import pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
plt.bar(x,y,width=0.5)
plt.show()
```

```python
x=[1,2,3,4,5]
y=[6,7,8,9,10]
c=['r','g','y','b','g']
plt.bar(x,y,color=c)
plt.show()
```

# Bar Plot

```python
x=['Python','C','C++','Java']
y=[85,70,60,80]
z=[44,55,30,60]
plt.xlabel('Language',fontsize=15)
plt.ylabel('No',fontsize=15)
plt.title('Dream',fontsize=15)
plt.bar(x,y,width=0.5,color='g',align='edge',edgecolor='b',label='Dream')
plt.bar(x,z,width=0.5,color='b',align='edge',edgecolor='b',label='Dream1')
plt.legend()
plt.show()
```
