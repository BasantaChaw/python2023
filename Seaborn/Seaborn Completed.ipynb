# Seaborn SCatter Plot

```python
t=sb.load_dataset('titanic')
t
```

```python
sb.scatterplot(x='age',y='fare',data=t)
```

```python
sb.scatterplot(x='age',y='fare',data=t,hue='sex')
```

```python
sb.scatterplot(x='age',y='fare',data=t,hue=t.sex,style='who',size='who')
```

```python
plt.figure(figsize=(16,10))
sb.scatterplot(x='age',y='fare',data=t,hue=t.sex,style='who',size='who',sizes=(100,200))
```

```python
plt.figure(figsize=(16,9))
sb.scatterplot(x='who',y='fare',data=t,hue=t.alive,style='alive',size='who',sizes=(100,400),palette='inferno_r',alpha=0.7)
```
